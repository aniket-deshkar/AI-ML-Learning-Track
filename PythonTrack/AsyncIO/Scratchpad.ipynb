```python
import asyncio

async def main():
    task = asyncio.create_task(calling_main())
    print("Main Function A")
    await asyncio.sleep(1)
    print("Main Function 2")
    return_value = await task
    print(return_value) #Future in Python

async def calling_main():
    print("Statement 1")
    await asyncio.sleep(2)
    print("Statement 2")
    return "EOF"

await main()
```
