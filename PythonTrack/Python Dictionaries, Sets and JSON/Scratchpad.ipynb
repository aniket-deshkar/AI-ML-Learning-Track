```python
def add(a,b):
    '''Addition of two numbers'''
    c =a+b
    return "A:",a,"B:",b,"Sum:",c
print(add.__doc__)
add(25,52)

```

```python
def mul(a,b=10):
    '''Multiplication of two numbers'''
    c =a*b
    return "A:",a,"B:",b,"Output:",c
print(mul.__doc__)
mul(10)

```

```python
def nums(*args):
    return args
print(nums(1,2,4))
```

```python
def dict_kwargs(**kwargs):
    return kwargs
print(dict_kwargs(name="Aniket"))
```

```python
def mixed_args(age,*hobby,name="Jack",**role):
    print("Age:",age)
    print("Hobby:",hobby)
    print("Name:",name)
    print("Role:",role)
    return age,name,*hobby,role
output = mixed_args(25,("Cricket"),role="Engineer")
print(output)
```

```python
def calculate(*numbers):
    return sum(numbers)/len(numbers)
calculate(10,20,40,40)
```
