```python
import copy

names_list = ["john","Doe", "Australia","barcelona","Dubai", "Madagascar", "delhi"]
print(names_list)
names_list.sort()
print(names_list)
names_list.sort(key=str.lower)
print(names_list)
names_list.append("11")
print(names_list)
name_copy = names_list.copy()
print("Copy:",name_copy)
name_deep_copy = copy.deepcopy(name_copy)
print("DeepCopy:",name_deep_copy)
```
